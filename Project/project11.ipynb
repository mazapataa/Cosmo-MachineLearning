```python
from gravipy.tensorial import *
from sympy import *
```

```python
# we use spherical coordinates 
tau, r, theta, phi = symbols('tau r theta phi')
x = Coordinates('x', [tau, r, theta, phi])

a = Function('a')(tau)
Phi = Function('Phi')(tau, r, theta, phi)
Psi = Function('Psi')(tau, r, theta, phi)
```

```python
g_00 = a**2 * (1 + 2*Phi)
g_rr = -a**2 * (1 - 2*Psi)
g_psipsi = -a**2 * (1 - 2*Psi) * r**2
g_psi= -a**2 * (1 - 2*Psi) * r**2 * sin(theta)**2
```

```python

Metric = diag(g_00, g_rr, g_psipsi, g_psi)

#metric tensor
g = MetricTensor('g', x, Metric)
g(All,All)
```

```python
#Christoffel symbols
Ga = Christoffel('Ga', g)
Ga(All,All,All)
```

```python
Ri = Ricci('Ri', g)
G = Einstein('G', Ri)
G(All, All)
```

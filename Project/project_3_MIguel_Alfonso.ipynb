```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt

# Read the CLASS C_l output
data = np.loadtxt("/home/alfonsozapata/Documents/CLASS/class/output/explanatory00_cl.dat")
ell = data[:, 0].astype(int)
cl_tt = data[:, 1]  # mu K^2

# Convert from mu K^2 to K^2 for HEALPix
cl_tt_k2 = cl_tt * 1e-12

# Create full spectrum array up to desired lmax (must match CLASS output)
lmax = ell[-1]  # Maximum l in CLASS output
full_cl_tt = np.zeros(lmax + 1)
full_cl_tt[ell] = np.sqrt(cl_tt_k2)  # only K 

nside = 500
cmb_map = hp.synfast(full_cl_tt, nside=nside, lmax=2, new=True, verbose=False)


hp.mollview(cmb_map*1e6,  # Convert back to mu K for visualization
            title="CMB Temperature Map ($l=2$)",
            unit="$\mu$K")  
plt.show()
```

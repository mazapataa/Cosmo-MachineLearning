```python
from scipy.integrate import odeint 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ode1(x,y):
    ode = x + x*y
    return ode 

xvals = np.linspace(0,10,50)

sol1 = odeint(ode1,1.0,xvals)
```

```python


def ode2(x,y):
    ode = -3 -y/x
    return ode 

xvals2 = np.linspace(1.0,10,100)

sol2 = odeint(ode2,1,xvals2)
```

```python

def ode3(x,y):
    ode = 4*x - 2 * x*y
    return ode 



sol3 = odeint(ode3,10,xvals2)

plt.plot(xvals2,sol3)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

```python

def ode4(x,y):
    ode = y/x + np.tan(y/x)
    return ode 



sol4 = odeint(ode4,10,xvals2)

plt.plot(xvals2,sol4)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

```python

def ode5(x,y):
    ode = - y/x + 2*x**3 * y**4
    return ode 

xvals4 = np.linspace(0.001,2,100)

sol5 = odeint(ode5,1.0,xvals4)

plt.plot(xvals2,sol5)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

```python

def system(Y, x):
    y1, y2 = Y
    dy1_dx = y2
    dy2_dx = 2*y2 - y1 + np.exp(x)
    return [dy1_dx, dy2_dx]

# Define the initial conditions
y1_0 = 0  # Initial condition for y
y2_0 = 0  # Initial condition for dy/dx
Y0 = [y1_0, y2_0]



# Solve the system using odeint
solution = odeint(system, Y0, xvals2)

# Extract solutions
y1_solution = solution[:, 0]  # y values

plt.plot(xvals2, y1_solution)

```
